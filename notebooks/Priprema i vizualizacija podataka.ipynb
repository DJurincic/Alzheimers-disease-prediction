# Monotoni i nevažni atributi

Broj različitih vrijednosti koje pojedini atribut poprima možemo saznati funkcijom $nunique$.

```python
cs.nunique()
```

Atribut $ID$ je monotono rastući atribut čija vrijednost nam ne donosi značajne informacije tako da ga možemo izbaciti. Atribut $Hand$ poprima samo jednu vrijednost tako da nam ne donosi informacije.

```python
cs.drop(['ID', 'Hand'], axis=1, inplace=True)
```

```python
long.nunique()
```

Atributi $Subject$ $ID$ i $MRI$ $ID$ su identifikacijski atributi pa nam nisu potrebni. Atribut $Group$ nam nije potreban jer sadrži istu informaciju kao i atribut $CDR$, a atribut $CDR$ pokušavamo predvidjeti. Atribut $Hand$ poprima samo jednu vrijednost pa ga možemo izbaciti.

```python
long.drop(['Subject ID', 'MRI ID', 'Group', 'Hand', 'Visit', 'MR Delay'], axis=1, inplace=True)
```

# Nedostajući podatci

Moramo se pozabaviti i nedostajućim podacima. Postoji nekoliko načina na koje možemo riješiti taj problem, a konkretan način ovisi o atributu, njegovim vrijednostima i broju vrijednosti koje nedostaju.

```python
cs.isna().sum()
```

Atribut $Delay$ nam ne donosi puno informacija, a u većini redova nedostaje, tako da ćemo ga odbaciti.

```python
cs.drop(['Delay'], axis=1, inplace=True)
```

Atribut $CDR$ pokušavamo predvidjeti pa ćemo zbog toga odbaciti sve retke u kojima taj atribut nedostaje.

```python
cs=cs.loc[cs.CDR.notnull(), :]
```

Nedostajuće vrijednosti atributa $Educ$ i $SES$, koji predstavljaju stupanj obrazovanja i socio-ekonomski status, zamijeniti ćemo njihovim medijanima jer se radi o kategorijskom atributu za koji je razumno pretpostaviti srednju vrijednost.

```python
cs.loc[cs.Educ.isna(), 'Educ']=cs.loc[:,'Educ'].median()
cs.loc[cs.SES.isna(), 'SES']=cs.loc[:,'SES'].median()
```

Atribut $MMSE$ je broj bodova postignut na mini-mental state examinationu. Jedna od mana tog testa je pristranost protiv slabije obrazovanih ljudi jer test sadrži elemente jezičnog i matematičkog testiranja. Zbog toga bismo nedostajuće vrijednosti tog atributa trebali zamijeniti sa srednjom vrijednosti grupiranoj po razini obrazovanja.

Provjerimo postoji li razlika u $MMSE$ rezultatima s obzirom na stupanj obrazovanja.

```python
cs.groupby('Educ').MMSE.mean()
```

Rezultati potvrđuju da postoji razlika.

```python
cs['MMSE'] = cs.groupby('Educ')['MMSE'].transform(
    lambda x: x.fillna(np.mean(x)))
```

```python
cs.isna().sum()
```

Isti postupak ponavljamo na skupu $longitudinal$.

```python
long.isna().sum()
```

Nedostaju nam podaci za dva atributa s kojima smo se upoznali u prethodnom dijelu zadatka tako da ćemo ponoviti iste postupke kao i prije.

```python
long.loc[long.SES.isna(), 'SES']=long.loc[:,'SES'].median()
```

```python
long.groupby('EDUC').MMSE.mean()
```

```python
long['MMSE'] = long.groupby('EDUC')['MMSE'].transform(
    lambda x: x.fillna(np.mean(x)))
```

```python
long.isna().sum()
```

# Normalizacija vrijednosti i spajanje skupova

Vidimo da nam se skupovi podataka podudaraju u imenima stupaca, ali stupci $Educ$ imaju različite raspone. Zbog toga ćemo podatke iz skupa $longitudinal$ normalizirati na vrijednosti od 1 do 5.

```python
import math
#Mijenjanje imena stupca EDUC u Educ kako bi se imena u oba skupa poklapala
long.rename(columns={"EDUC":"Educ"}, inplace=True)

long['Educ']=((long['Educ']/long['Educ'].max())*5).apply(np.round)

long['Educ'].describe()
```

Sada nam se oba skupa podataka podudaraju u stupcima i rasponima vrijednosti pa ih možemo spojiti.

```python
combined= pd.concat([cs, long], axis=0)
```

```python
combined.describe()
```

Kombinirani skup podataka ćemo normalizirati Min-Max metodom.

```python
columns=list(combined.columns)
columns.remove("M/F")

for column in columns:
    combined[column]=(combined[column]-combined[column].min())/(combined[column].max()-combined[column].min())
combined.describe()
```

# Stršeći podaci

Stršeće podatke ili outliere najlakše uočavamo korištenjem box plot-a.

```python
columns=list(cs.columns)
columns.remove('M/F')

for atribute in columns:
    fig, ax=plt.subplots(1,2)
    
    fig.suptitle(atribute)    
    
    ax[0].boxplot(cs[atribute])
    ax[0].set_title("Cross-sectional")
    ax[1].boxplot(long[atribute])
    ax[1].set_title("Longitudinal")
```

Izgleda da u skupu podataka $cross$-$sectional$ nema outliera koji su prouzrokovani greškom u očitavanju ili upisivanju podataka jer su svi unutar razumnog raspona vrijednosti. U skupu podataka $longitudinal$ postoje značajniji outlieri u atributu $MMSE$, no također su unutar dozvoljenog raspona vrijednosti.

```python
columns=list(combined.columns)
columns.remove('M/F')

for atribute in columns:
    plt.boxplot(combined[atribute])
    plt.title(atribute)
    plt.figure()
```

U kombiniranom skupu podataka se ponovo pojavljuju outlieri u atributu $MMSE$.

# Promjena formata značajki

Atribut $M/F$ je tekstualnog oblika pa ga trebamo pretvoriti u brojčani korištenjem LabelEncoder-a.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(cs.loc[:,'M/F'])
cs.loc[:,'M/F'] = le.transform(cs.loc[:,'M/F'])


le = LabelEncoder()
le.fit(long.loc[:,'M/F'])
long.loc[:,'M/F'] = le.transform(long.loc[:,'M/F'])

le = LabelEncoder()
le.fit(combined.loc[:,'M/F'])
combined.loc[:,'M/F'] = le.transform(combined.loc[:,'M/F'])
```

Ove pretpostavke možemo grafički provjeriti crtanjem zasebnih grafove za pacijente koji imaju demenciju i koji nemaju demenciju, te pogledom na deskriptivnu statistiku.

```python
columns=["MMSE", "Age"]
demented=combined[combined["CDR"]>=0.5]
non_demented=combined[combined["CDR"]<0.5]

for column in columns:

    fig, ax=plt.subplots(1,2)
    
    fig.suptitle(column)    
    
    ax[0].hist(non_demented[column])
    ax[0].set_title("Non-demented")
    ax[1].hist(demented[column])
    ax[1].set_title("Demented")
```
